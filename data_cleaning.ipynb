```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_column = null_var[null_var>10].keys()
drop_column
```

```python
df_drop_column = df.drop(columns=drop_column)
df_drop_column.shape
```

```python
sns.heatmap(df_drop_column.isnull())
```

```python
df_drop_rows = df_drop_column.dropna()
df_drop_rows.shape
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df_drop_rows.isnull())
```

```python
df_drop_rows.isnull().sum().sum()
```

```python
df_drop_rows.select_dtypes(include='object').columns
```

```python
pd.concat([df['areaWithType'].value_counts()/df.shape[0]*100,
          df_drop_rows['areaWithType'].value_counts()/df_drop_rows.shape[0]*100],
          axis=1, keys=['areaWithType_unclean', 'areaWithType_clean'])
```

```python
def cat_dist(var):
  return pd.concat([df[var].value_counts()/df.shape[0]*100,
          df_drop_rows[var].value_counts()/df_drop_rows.shape[0]*100],
          axis=1, keys=[var+'_unclean', var+'_clean'])
```

```python
cat_dist('furnishing')
```

```python
plt.figure(figsize=(8,8))
for i,var in enumerate(df_drop_rows.select_dtypes(include='object').columns):
  plt.subplot(3,3,i+1)
  sns.histplot(df_drop_rows[var], bins=20)
  # plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12, 12))  # Adjusted the figure size for better spacing
for i, var in enumerate(df_drop_rows.select_dtypes(include='object').columns):
    plt.subplot(3, 3, i + 1)
    sns.countplot(y=var, data=df_drop_rows)  # Count plot for categorical variables
    plt.title(f'Distribution of {var}')
    plt.tight_layout()
```

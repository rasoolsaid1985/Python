```python
winners_list=dataset['Winners']
winners_list
```

```python
dataset[dataset['Winners']=='Rain']
```

```python
teams=(set)((list)(dataset['1st Team'].unique())+(list)(dataset['2nd Team'].unique()))
teams=list(teams)
print(teams)
```

```python
# teams = pd.concat([dataset['1st Team'], dataset['2nd Team']]).drop_duplicates().tolist()
# print(teams)
```

```python
number_of_wins=[]
for team in teams:
    c=0
    for winner in winners_list:
        if team==winner:
            c+=1
    number_of_wins.append(c)
number_of_wins
```

```python
points_earned={}
for team in teams:
    c=0
    for winner in winners_list:
        if team==winner:
            c+=2
    points_earned[team]=c
team_affected_by_rain=list(dataset[dataset['Winners']=='Rain']['1st Team'])+list(dataset[dataset['Winners']=='Rain']['2nd Team'])
for team in team_affected_by_rain:
    points_earned[team]+=1
points_earned
points_earned
```

```python
c=0
print("Team name     Number of wins      Points")
while(c<20):
    print(teams[c],'\t',number_of_wins[c],'\t',points_earned[teams[c]])
    c+=1
```

```python
data = {
    'Team': teams,
    'Number of Wins': number_of_wins,
    'Points Earned': points_earned.values()
}

# Create the DataFrame
points_table = pd.DataFrame(data,index=range(1,21))
points_table
```

```python
top_batting_scorers = dataset[dataset['Top Scorer'] != 'Rain']['Top Scorer'].value_counts().head(7) # Excluding Rain from 'Top Scorer'
top_batting_scorers_names=list(pd.DataFrame(top_batting_scorers).index)
top_batting_scorers_names
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=dataset,order=top_batting_scorers.index,y="Top Scorer")
plt.title("Top 7 Scorers")
plt.xlabel("Number of Matches")
plt.ylabel("Players")
plt.show()
```

```python
dataset[dataset['Top Scorer']=='Quinton de Kock']['Highest Score'].values
```

```python
batsman_runs=[]
for name in top_batting_scorers_names:
    scores=dataset[dataset['Top Scorer']==name]['Highest Score'].values
    sum_of_runs=0
    for i in scores:
        sum_of_runs+=int(i)
    batsman_runs.append(sum_of_runs)

data={
    'Name':top_batting_scorers_names,
    'Runs Scored':batsman_runs
}
df=pd.DataFrame(data,index=range(1,8))
# Sort the DataFrame in descending order based on 'Score'
df= df.sort_values(by='Runs Scored', ascending=False)
df
```

```python
# Exclude Rain from the player of the match
players_of_matches = dataset[dataset['Player Of The Match'] != 'Rain']['Player Of The Match'].value_counts().head(7)
plt.figure(figsize=(8, 4))
sns.barplot(x=players_of_matches.values, y=players_of_matches.index)
plt.title('Top 7 Players of the Match')
plt.xlabel('Number of Matches')
plt.ylabel('Players')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=dataset, x='Method')
# Annotate each bar with its count
for p in plt.gca().patches:
    plt.gca().annotate(f'\n{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                       textcoords='offset points')
plt.title('Winning Method Analysis')
plt.xlabel('Method')
plt.ylabel('Count')
plt.show()
```
